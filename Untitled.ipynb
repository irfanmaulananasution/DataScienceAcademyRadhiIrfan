### Benerin Outlier (error input data IPK)

```python
#fungsi untuk merewrite IPK yang koma nya tertinggal (misal IPK 400 harusnya 4.00)
def correct_IPK(ipk):
    if (ipk < 4.00):
        return ipk
    elif (ipk < 40.00):
        return ipk/10.00
    elif (ipk < 400.00):
        return ipk/100.00
```

```python
dat['IPK'] = dat['IPK'].apply(correct_IPK)
```

# B. Visualization

Fungsi penghapus outlier

```python
def rub(q1, q3, max): #dat["something"]
    result = (5*q3 - 3*q1)/2
    if (result > max):
        return max
    else :
        return result

def rlb(q1, q3, min):
    result = (5*q1 - 3*q3)/2
    if (result < min):
        return min
    else :
        return result

def removeOutlier(data, min, max, q1, q2, q3):
    tmpRub = rub(q1, q3, max)
    tmpRlb = rlb(q1, q3, min)
    if( data > tmpRub):
        return q2
    elif( data < tmpRlb):
        return q2
    else :
        return data
        
```

# C. Korelasi

```python
def convertPOrtu(datum):
    if ( datum == ">10 juta"):
        return 4
    elif ( datum == "8-10 juta"):
        return 3
    elif ( datum == "5-8 juta"):
        return 2
    elif ( datum == "3-5 juta"):
        return 1
    else:
        return 0
```

```python
dat_num = dat[['IPK', 'Semester', 'Jarak dari rumah ke kampus (kalau PP) dalam jarak kilometer (km)', 'Pendapatan tiap bulan (termasuk uang saku dari orang tua)']]
```

```python
dat_num['Penghasilan Orang Tua'] = dat['Penghasilan Orang Tua'].apply(convertPOrtu)
dat_num['Jenis Kelamin'] = dat['Jenis Kelamin'].apply(lambda x: 1 if x == 'Pria' else 0)
dat_num['Tempat Tinggal'] = dat['Tempat Tinggal'].apply(lambda x: 1 if x == 'kos' else 0)
dat_num['Apakah Anda kuliah sambil bekerja?'] = dat['Apakah Anda kuliah sambil bekerja?'].apply(lambda x: 1 if x == 'Ya' else 0)
dat_num['Apakah Anda mengikuti organisasi?'] = dat['Apakah Anda mengikuti organisasi?'].apply(lambda x: 1 if x == 'Ya' else 0)
```

```python
correlation = dat_num.corr(method='spearman')
sns.heatmap(correlation, annot=True)
plt.savefig('10 Korelasi Spearman.png')
```

```python
correlation = dat_num.corr(method='pearson')
sns.heatmap(correlation, annot=True)
plt.savefig('10 Korelasi Pearson.png')
```

```python
correlation = dat_num.corr(method='kendall')
sns.heatmap(correlation, annot=True)
plt.savefig('10 Korelasi Kendall.png')
```
